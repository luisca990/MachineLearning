```python
n_components=11
df4=df3.loc[:,['SalePrice','OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd','New_y_Var']]
# Do the PCA.
pca = PCA(n_components=n_components)
reduced = pca.fit_transform(df4)

# Append the principle components for each entry to the dataframe
for i in range(0, n_components):
    df4['PC' + str(i + 1)] = reduced[:, i]

display(df4.head())

# Do a scree plot
ind = np.arange(0, n_components)
(fig, ax) = plt.subplots(figsize=(8, 6))
sns.pointplot(x=ind, y=pca.explained_variance_ratio_)
ax.set_title('Scree plot')
ax.set_xticks(ind)
ax.set_xticklabels(ind)
ax.set_xlabel('Component Number')
ax.set_ylabel('Explained Variance')
plt.show()

# Show the points in terms of the first two PCs
g = sns.lmplot('PC1',
               'PC2',
               hue='New_y_Var',data=df4,
               fit_reg=False,
               scatter=True,
               size=7)

plt.show()

# Plot a variable factor map for the first two dimensions.
(fig, ax) = plt.subplots(figsize=(20, 20))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Start the arrow at the origin
             pca.components_[0, i],  #0 for PC1
             pca.components_[1, i],  #1 for PC2
             head_width=0.1,
             head_length=0.1)

    plt.text(pca.components_[0, i] + 0.05,
             pca.components_[1, i] + 0.05,
             df4.columns.values[i])


an = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.cos(an), np.sin(an))  # Add a unit circle for scale
plt.axis('equal')
ax.set_title('Variable factor map')
plt.show()
```

### De acuerdo a los resultados de la PCA se evidencia como las variables relacionadas con el área de las casas son las más relevantes. La variable "TotalBsmtSF" que corresponde al área total del sótano es la más importante, seguida de la variable " 1sFirSF" que corresponde al área del primer piso y la tercera variable "GrLivArea" que corresponde al área total habitable de la casa.

## 2. Random Forest

```python
y_train = df5['New_y_Var'].astype("int")
x_train = df5.loc[:,['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd']]
from sklearn.ensemble import RandomForestClassifier
Random_F = RandomForestClassifier(n_estimators = 50,max_depth=8, random_state=0)
Random_Fit = Random_F.fit(x_train, y_train)
```

```python
Matrix=confusion_matrix(y_train, Random_Fit.predict(x_train))
```

```python
# get importance
from matplotlib import pyplot
importance = Random_Fit.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance

plt.barh(x_train.columns, Random_Fit.feature_importances_)
```

```python
#### Accuracy

Accuracy=(Matrix[0][0]+Matrix[1][1])/(Matrix[0][0]+Matrix[0][1]+Matrix[1][0]+Matrix[1][1])
print("El Accuracy es: ", Accuracy)

#### Especificidad


Especificidad=(Matrix[0][0])/(Matrix[0][0]+Matrix[0][1])
print("La especificidad es:", Especificidad)

### Sensibilidad

sensitividad=(Matrix[1][1])/(Matrix[1][1]+Matrix[1][0])
print("La sensibilidad es: ",sensitividad)

### Valor predictivo positivo

PredictivoP=Matrix[1][1]/(Matrix[0][1]+Matrix[1][1])
print("El valor de predictivos positivos es: ",PredictivoP)

### Valor predictivo negativo

PredictivoN=Matrix[0][0]/(Matrix[0][0]+Matrix[1][0])
print("El valor de predictivos negativos es: ",PredictivoN)

MetricasRandomF=[Accuracy, Especificidad, sensitividad,PredictivoP, PredictivoN]
```

#### En conclusión es un modelo bueno en todas sus métricas, mejorando con respecto árboles de decisión en todas las métricas. La métrica de valores positivos es del 95%, siendo capaz de predecir en el 95% de los casos de manera correcta en el rango de precio de venta de las casas con valor inferior 214.000 USD y un 99% de las casas con valor superior a 214.000 USD

#### Inicialmente se pasaron las 10 variables, sin embargo al obtner el gráfico de importancia se quita la variable baños completos sobre rasante (FullBath) notando una leve mejoria en las métricas del modelo.

### 3. Gradient Boosting

```python
y_train = df5['New_y_Var'].astype("int")
x_train = df5.loc[:,['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'YearBuilt', 'YearRemodAdd']]

from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
        max_depth=1, random_state=0).fit(x_train, y_train)
```

```python
#### Accuracy

Accuracy=(MatrixG[0][0]+MatrixG[1][1])/(MatrixG[0][0]+MatrixG[0][1]+MatrixG[1][0]+MatrixG[1][1])
print("El Accuracy es: ", Accuracy)

#### Especificidad


Especificidad=(MatrixG[0][0])/(MatrixG[0][0]+MatrixG[0][1])
print("La especificidad es:", Especificidad)

### Sensibilidad

sensitividad=(MatrixG[1][1])/(MatrixG[1][1]+MatrixG[1][0])
print("La sensibilidad es: ",sensitividad)

### Valor predictivo positivo

PredictivoP=MatrixG[1][1]/(MatrixG[0][1]+MatrixG[1][1])
print("El valor de predictivos positivos es: ",PredictivoP)

### Valor predictivo negativo

PredictivoN=MatrixG[0][0]/(MatrixG[0][0]+MatrixG[1][0])
print("El valor de predictivos negativos es: ",PredictivoN)

MetricasRandomG=[Accuracy, Especificidad, sensitividad,PredictivoP, PredictivoN]
```

#### En conclusión es un modelo bueno en todas sus métricas, mejorando con respecto árboles de decisión en casi todas las métricas excepto Sensibilidad, pero no mejoro con respecto a RandomForest. La métrica de valores positivos es del 89%, siendo capaz de predecir en elb 89% de los casos de manera correcta en el rango de precio de venta de las casas con valor inferior 214.000 USD y un 97% de las casas con valor superior a 214.000 USD

#### Inicialmente se pasaron las 10 variables, sin embargo se hicieron pruebas quitando las variables de baños completos sobre rasante (FullBath) y total de habitaciones sobre rasante (no incluye baños) ( TotRmsAbvGrd) notando mejoría en las métricas.
