### 4. Ubicar las columnas que contienen valores vacios usando la función Unique()

```python
import numpy as np
def isNaN(num):
    return num != num
array2=[]
for col in df:
  print(col,"\n")
  array=df[col].unique()
  print(array)
  for dato in array:
    if(isNaN(dato)):
      array2.append(col)
      continue
print(array2)


 
```

### 9. PCA.

```python
import numpy as np
from sklearn.decomposition import PCA
columnsUse=['SalePrice','New_y_Var','OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd']
df3=df2.loc[:,columnsUse]
df3.head()
```

```python
pca = PCA(n_components=10)
pca.fit(df3.loc[:,['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd']])
pca.components_
```

```python
principalComponents = pca.fit_transform(df3.loc[:,['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd']])
INDIVID = pd.DataFrame(principalComponents)
INDIVID
```

```python
n_components=10
df4=df3.loc[:,['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd','New_y_Var']]
# Do the PCA.
pca = PCA(n_components=n_components)
reduced = pca.fit_transform(df4)

# Append the principle components for each entry to the dataframe
for i in range(0, n_components):
    df4['PC' + str(i + 1)] = reduced[:, i]

display(df4.head())

# Do a scree plot
ind = np.arange(0, n_components)
(fig, ax) = plt.subplots(figsize=(8, 6))
sns.pointplot(x=ind, y=pca.explained_variance_ratio_)
ax.set_title('Scree plot')
ax.set_xticks(ind)
ax.set_xticklabels(ind)
ax.set_xlabel('Component Number')
ax.set_ylabel('Explained Variance')
plt.show()

# Show the points in terms of the first two PCs
g = sns.lmplot('PC1',
               'PC2',
               hue='New_y_Var',data=df4,
               fit_reg=False,
               scatter=True,
               size=7)

plt.show()

# Plot a variable factor map for the first two dimensions.
(fig, ax) = plt.subplots(figsize=(20, 20))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Start the arrow at the origin
             pca.components_[0, i],  #0 for PC1
             pca.components_[1, i],  #1 for PC2
             head_width=0.1,
             head_length=0.1)

    plt.text(pca.components_[0, i] + 0.05,
             pca.components_[1, i] + 0.05,
             df4.loc[:,['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd','New_y_Var']].columns.values[i])


an = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.cos(an), np.sin(an))  # Add a unit circle for scale
plt.axis('equal')
ax.set_title('Variable factor map')
plt.show()
```
